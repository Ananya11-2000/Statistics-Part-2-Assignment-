```python
#Q-1 What is hypothesis testing in statistics?
#A-1 It is a formal method used to make inferences or decisions about a population based on sample data.
#It helps determine whether there is enough evidence in a sample to support a specific belief (or hypothesis) about the population.
```

```python
#Q-2 What is the null hypothesis, and how does it differ from the alternative hypothesis?
#A-2 Null hypothesis is a default or starting assumption that there is no effect, no difference, or no relationship in the population.
# Its main goal is to challenge or test this assumption using sample data.
# Whereas alternative hypothesis is the statement you are trying to find evidence for. It suggests that there is an effect, a difference, or a relationship.
```

```python
#Q-3 What is the significance level in hypothesis testing, and why is it important?
#A-3 The significance level, denoted by α (alpha), is the threshold you set before conducting a hypothesis test to determine how much risk of a false positive (Type I error) you're willing to accept.
# It is important because it Controls Error Risk, Guides Decision-Making and Balances Sensitivity and Specificity.
# It plays a crucial role in maintaining the integrity of statistical conclusions.
```

```python
#Q-4 What does a P-value represent in hypothesis testing?
#A-4 A p-value is the probability of obtaining results at least as extreme as the observed ones, assuming the null hypothesis (H₀) is true.
# In simple words it tells us how likely our results are just due to random chance.
```

```python
#Q-5 How do you interpret the P-value in hypothesis testing?
#A-5 The p-value helps you determine whether to reject the null hypothesis (H₀) based on the evidence from your data.
# It is the probability of observing a test statistic as extreme or more extreme than what was actually observed, assuming H₀ is true.
```

```python
#Q-6 What are Type 1 and Type 2 errors in hypothesis testing?
#A-6 n hypothesis testing, We make a decision based on sample data, and like any decision, there’s a risk of being wrong.
# These two types of wrong decisions are called Type I and Type II errors.
# In Type I Error (False Positive): We Reject the null hypothesis (H₀) when it is actually true means we say there is an effect when there isn’t.
# For eg. A medical test says a person has a disease when they don’t have it.
# In Type II Error (False Negative): We Fail to reject the null hypothesis (H₀) when the alternative hypothesis (H₁) is actually true.
# For eg. A test says a person is healthy when they actually have the disease.
```

```python
#Q-7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
#A-7 One-Tailed Test: Tests for an effect in one specific direction.
# Its alternative hypothesis (H₁): Uses greater than (>) or less than (<).
# Two-Tailed Test: Tests for an effect in either direction (greater or smaller).
# Its alternative hypothesis (H₁): Uses not equal to (≠).
```

```python
#Q-8 What is the Z-test, and when is it used in hypothesis testing?
#A-8 The Z-test is a statistical method used to determine whether there's a significant difference between a sample statistic and
#a known population parameter—when certain conditions are met.
# It is used :-  when the population standard deviation (σ) is known and the sample size is large (typically n ≥ 30).
```

```python
#Q-9 How do you calculate the Z-score, and what does it represent in hypothesis testing?
#A-9 A Z-score measures how many standard deviations a data point (or sample mean) is from the population mean under the null hypothesis.
# Z-score formula : Z = sample mean - Population mean under the null hypothesis (H₀)/ Population standard deviation/Sample size
# A Z-score tells you how far your sample statistic is from the population mean, in standard deviation units.
# It allows you to compare your result to the standard normal distribution (Z-distribution) to determine the p-value and test significance.
```

```python
#Q-10 What is the T-distribution, and when should it be used instead of the normal distribution?
#A-10 The t-distribution (Student’s t-distribution) is a probability distribution used in statistics.
# It is used when we are working with small sample sizes (usually n < 30) and population standard deviation (σ) is unknown.
```

```python
#Q-11 What is the difference between a Z-test and a T-test?
#A-11 Z-test :- It is used when Population standard deviation (σ) is known, sample size is Large (n ≥ 30) or known σ,
# uses Standard normal distribution (Z) and have fixed bell curve shape.
# T-test :- It is used when Population standard deviation (σ) is unknown, sample size is Small (n < 30) or unknown σ,
# uses T-distribution (with heavier tails) and have Wider shape, changes shape depending on degrees of freedom.
```

```python
#Q-12 What is the T-test, and how is it used in hypothesis testing?
#A-12 The T-test is a statistical method used to determine whether there is a significant difference between means
#especially when the sample size is small and the population standard deviation is unknown.
# It is used when we're comparing sample means (one group or between two groups), population standard deviation (σ) is unknown and sample size is small (n < 30) or the population isn't large.
```

```python
#Q-13 What is the relationship between Z-test and T-test in hypothesis testing?
#A-13 The Z-test and T-test are both statistical methods used in hypothesis testing
#to evaluate whether there is a significant difference between a sample statistic and a population parameter (usually the mean).
#While they serve the same purpose, they are used in different situations depending on the information available and sample size.
```

```python
#Q-14 What is a confidence interval, and how is it used to interpret statistical results?
#A-14 A confidence interval (CI) is a range of values that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence (commonly 95%).
# A 95% confidence interval means that if we repeated the sampling process many times, 95% of the resulting intervals would contain the true population value.
# Confidence Interval = Point Estimate ± Margin of Error
```

```python
#Q-15 What is the margin of error, and how does it affect the confidence interval?
#A-15 The margin of error is the amount you're allowing for potential variation or uncertainty in your estimate.
#It defines how wide the confidence interval (CI) is around the sample statistic (like the mean or proportion).
# Margin of Error = Critical Value × Standard Error
# The confidence interval is calculated as: Confidence Interval = Point Estimate ± Margin of Error
#So,
# A larger margin of error → wider interval → less precision and A smaller margin of error → narrower interval → more precision.
```

```python
#Q-16 How is Bayes' Theorem used in statistics, and what is its significance?
#A-16 Bayes' Theorem is a foundational concept in statistics that provides a way to update the probability of a hypothesis based on new evidence.
# Bayes' Theorem: The Formula P(A∣B) = P(B∣A)⋅P(A)/P(B)
# Uses of bayes' theorem in statistics:
#Updating Beliefs: Bayes' Theorem is used to update probabilities as more data becomes available. For example, if a doctor has a prior belief about the likelihood of a disease, they can use test results (evidence) to update that belief.
#Bayesian Inference: This is a statistical method that applies Bayes' Theorem to estimate the distribution of a parameter. Unlike classical (frequentist) inference, Bayesian methods treat parameters as random variables with distributions
#Decision Making Under Uncertainty: In fields like economics, medicine, and artificial intelligence, Bayes' Theorem helps in making optimal decisions by weighing probabilities and costs.
#Machine Learning and AI: Algorithms such as Naive Bayes classifiers use Bayes' Theorem to classify data. It’s also central to Bayesian networks, which model complex probabilistic relationships
#Hypothesis Testing: In Bayesian hypothesis testing, you compute the posterior probabilities of different hypotheses rather than relying on p-values.

#Significance of Bayes' Theoem:
#Subjective Probabilities: Bayes' Theorem allows incorporation of prior knowledge or expert opinion, which is valuable when data is sparse or incomplete.
#Flexibility: Bayesian methods can handle complex models and are especially powerful in hierarchical modeling.
#Interpretability: The results from Bayesian analysis (like credible intervals) often have more intuitive interpretations than frequentist confidence intervals.
```

```python
#A-3
def bernoulli_pmf(x, p):
    """
    Calculates the PMF (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The value (0 or 1)
    p (float): The probability of success (0 <= p <= 1)

    Returns:
    float: The probability P(X = x)
    """
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0.0

print("P(X=0):", bernoulli_pmf(0, 0.3))
print("P(X=1):", bernoulli_pmf(1, 0.3))
print("P(X=2):", bernoulli_pmf(2, 0.3))
```

```python
#A-9
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculates Z-scores from a dataset and plots their histogram.

    Parameters:
    data (array-like): Input dataset (list or NumPy array)
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7, density=True)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axvline(x=0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

if __name__ == "__main__":
    sample_data = np.random.normal(loc=50, scale=10, size=1000)
    z_scores = calculate_and_plot_zscores(sample_data)
```

```python
#A-10
import numpy as np
import matplotlib.pyplot as plt

def simulate_clt(distribution, sample_size=30, num_samples=1000):
    """
    Simulate CLT using a specified distribution.

    Parameters:
    - distribution: function to generate random data (e.g., np.random.exponential)
    - sample_size: size of each sample drawn
    - num_samples: number of sample means to calculate
    """
    sample_means = []

    for _ in range(num_samples):
        sample = distribution(size=sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    plt.hist(sample_means, bins=40, density=True, alpha=0.7, edgecolor='black')
    plt.title(f'CLT Simulation (n={sample_size}, samples={num_samples})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

if __name__ == "__main__":
    simulate_clt(lambda size: np.random.exponential(scale=1.0, size=size), sample_size=30, num_samples=1000)
```

```python
#A-11
import numpy as np
import matplotlib.pyplot as plt

def simulate_clt_normal(mu=50, sigma=10, sample_size=30, num_samples=1000):
    """
    Simulate CLT using a normal distribution and plot the distribution of sample means.

    Parameters:
    - mu: population mean
    - sigma: population standard deviation
    - sample_size: size of each sample
    - num_samples: number of samples to draw
    """
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    plt.hist(sample_means, bins=40, density=True, edgecolor='black', alpha=0.7)
    plt.title(f'Sampling Distribution of the Mean\n(Normal Population: μ={mu}, σ={sigma}, n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axvline(np.mean(sample_means), color='red', linestyle='--', label='Mean of Sample Means')
    plt.legend()
    plt.show()

    print(f"Mean of sample means ≈ {np.mean(sample_means):.2f}")
    print(f"Standard deviation of sample means ≈ {np.std(sample_means):.2f} (should ≈ σ/√n = {sigma/np.sqrt(sample_size):.2f})")

if __name__ == "__main__":
    simulate_clt_normal(mu=50, sigma=10, sample_size=30, num_samples=1000)
```

```python
#A-14
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculates the Z-score for a given data point.

    Parameters:
    - x: the data point
    - mean: mean of the dataset
    - std: standard deviation of the dataset

    Returns:
    - Z-score
    """
    return (x - mean) / std

def plot_z_score(z):
    """
    Plots the standard normal distribution and highlights the Z-score area.

    Parameters:
    - z: calculated Z-score
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, y, where=(x <= z), color='skyblue', alpha=0.6, label=f'Area ≤ Z ({z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.axvline(0, color='black', linestyle='--', linewidth=0.8)

    plt.title('Z-score on the Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    x = 75
    mean = 70
    std = 10

    z = calculate_z_score(x, mean, std)
    print(f"Z-score for x = {x} (mean = {mean}, std = {std}) is: {z:.2f}")

    plot_z_score(z)
```

```python
#A-15
import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, sigma, alpha=0.05):
    """
    Perform a Z-test for the population mean with known population std deviation.

    Parameters:
    - sample: array-like sample data
    - mu_0: hypothesized population mean
    - sigma: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - Z-score, p-value, decision
    """
    sample = np.array(sample)
    n = len(sample)
    x_bar = np.mean(sample)

    z = (x_bar - mu_0) / (sigma / np.sqrt(n))

    p_value = 2 * (1 - norm.cdf(abs(z)))

    reject = p_value < alpha
    decision = "Reject H0" if reject else "Fail to reject H0"

    print(f"Sample Mean = {x_bar:.2f}")
    print(f"Hypothesized Mean = {mu_0}")
    print(f"Z-Statistic = {z:.4f}")
    print(f"P-Value = {p_value:.4f}")
    print(f"Decision: {decision}")

    return z, p_value, decision

if __name__ == "__main__":
    sample_data = [72, 68, 75, 70, 74, 69, 71, 73, 67, 76]
    population_std = 4
    hypothesized_mean = 70
    alpha = 0.05

    z_test(sample_data, mu_0=hypothesized_mean, sigma=population_std, alpha=alpha)
```

```python
#A-16
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates and prints the confidence interval for the sample mean using the t-distribution.

    Parameters:
    - data: array-like sample data
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - Tuple (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident the true population mean lies between {lower:.2f} and {upper:.2f}.")

    return (lower, upper)

if __name__ == "__main__":
    sample_data = [72, 68, 75, 70, 74, 69, 71, 73, 67, 76]
    confidence_interval(sample_data, confidence=0.95)
```

```python
#A-17
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def generate_and_ci(mu=100, sigma=15, n=30, confidence=0.95):
    """
    Generates data from a normal distribution, calculates the confidence interval,
    and interprets the result.

    Parameters:
    - mu: mean of the normal distribution
    - sigma: standard deviation of the distribution
    - n: sample size
    - confidence: confidence level (default = 0.95)
    """
    data = np.random.normal(loc=mu, scale=sigma, size=n)

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)
    df = n - 1

    t_crit = stats.t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_crit * standard_error
    lower = sample_mean - margin_of_error
    upper = sample_mean + margin_of_error

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval for Mean: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")


generate_and_ci(mu=100, sigma=15, n=30, confidence=0.95)
```

```python
#A-1
from math import sqrt
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05, alternative='two-sided'):
    """
    Perform a Z-test comparing a sample mean to a known population mean.

    Parameters:
    - sample_mean: Mean of the sample
    - population_mean: Known population mean (null hypothesis mean)
    - population_std: Known population standard deviation
    - sample_size: Number of observations in the sample
    - alpha: Significance level (default 0.05)
    - alternative: 'two-sided', 'less', or 'greater' for test type

    Returns:
    - z_stat: Calculated Z-statistic
    - p_value: Corresponding p-value
    - interpretation: Text interpretation of the test
    """
    std_error = population_std / sqrt(sample_size)

    z_stat = (sample_mean - population_mean) / std_error

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    if p_value < alpha:
        interpretation = f"Reject the null hypothesis at alpha={alpha}. There is significant evidence that the sample mean is different from the population mean."
        if alternative == 'greater':
            interpretation = f"Reject the null hypothesis at alpha={alpha}. There is significant evidence that the sample mean is greater than the population mean."
        elif alternative == 'less':
            interpretation = f"Reject the null hypothesis at alpha={alpha}. There is significant evidence that the sample mean is less than the population mean."
    else:
        interpretation = f"Fail to reject the null hypothesis at alpha={alpha}. There is not enough evidence to conclude that the sample mean is different from the population mean."
        if alternative == 'greater':
            interpretation = f"Fail to reject the null hypothesis at alpha={alpha}. There is not enough evidence to conclude that the sample mean is greater than the population mean."
        elif alternative == 'less':
            interpretation = f"Fail to reject the null hypothesis at alpha={alpha}. There is not enough evidence to conclude that the sample mean is less than the population mean."

    return z_stat, p_value, interpretation


if __name__ == "__main__":
    sample_mean = 105
    population_mean = 100
    population_std = 15
    sample_size = 30
    alpha = 0.05
    alternative = 'two-sided'

    z_stat, p_value, interpretation = z_test(sample_mean, population_mean, population_std, sample_size, alpha, alternative)
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(interpretation)
```

```python
#A-2
import numpy as np
from scipy.stats import norm
from math import sqrt

def simulate_and_test(pop_mean, pop_std, sample_size, alpha=0.05, alternative='two-sided'):
    """
    Simulate random sample data and perform a Z-test comparing sample mean to population mean.

    Parameters:
    - pop_mean: Population mean (null hypothesis mean)
    - pop_std: Population standard deviation (assumed known)
    - sample_size: Number of samples to generate
    - alpha: Significance level
    - alternative: 'two-sided', 'greater', or 'less'

    Returns:
    - sample_mean: Mean of simulated sample
    - z_stat: Calculated Z-statistic
    - p_value: p-value from the test
    - interpretation: Result interpretation string
    """

    sample = np.random.normal(loc=pop_mean, scale=pop_std, size=sample_size)
    sample_mean = np.mean(sample)

    std_error = pop_std / sqrt(sample_size)

    z_stat = (sample_mean - pop_mean) / std_error

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    if p_value < alpha:
        interpretation = f"Reject the null hypothesis at alpha={alpha}. Sample mean ({sample_mean:.3f}) is significantly different from population mean ({pop_mean})."
        if alternative == 'greater':
            interpretation = f"Reject null hypothesis: sample mean is significantly greater than population mean."
        elif alternative == 'less':
            interpretation = f"Reject null hypothesis: sample mean is significantly less than population mean."
    else:
        interpretation = f"Fail to reject the null hypothesis at alpha={alpha}. No significant difference between sample mean ({sample_mean:.3f}) and population mean ({pop_mean})."
        if alternative == 'greater':
            interpretation = f"Fail to reject null hypothesis: sample mean is not significantly greater than population mean."
        elif alternative == 'less':
            interpretation = f"Fail to reject null hypothesis: sample mean is not significantly less than population mean."

    return sample_mean, z_stat, p_value, interpretation


if __name__ == "__main__":
    pop_mean = 50
    pop_std = 10
    sample_size = 40
    alpha = 0.05
    alternative = 'two-sided'

    sample_mean, z_stat, p_value, interpretation = simulate_and_test(pop_mean, pop_std, sample_size, alpha, alternative)

    print(f"Simulated sample mean: {sample_mean:.3f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(interpretation)
```

```python
#A-3
from math import sqrt
from scipy.stats import norm

def one_sample_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test to compare sample mean to population mean.

    Parameters:
    - sample_mean: Mean of the sample
    - population_mean: Known population mean
    - population_std: Known population standard deviation
    - sample_size: Number of observations in the sample
    - alpha: Significance level (default 0.05)
    - alternative: 'two-sided', 'greater', or 'less'

    Returns:
    - z_stat: Z-test statistic
    - p_value: P-value corresponding to the test statistic
    - conclusion: Interpretation of the hypothesis test result
    """

    std_error = population_std / sqrt(sample_size)

    z_stat = (sample_mean - population_mean) / std_error

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    if p_value < alpha:
        if alternative == 'two-sided':
            conclusion = f"Reject null hypothesis: sample mean is significantly different from population mean at alpha={alpha}."
        elif alternative == 'greater':
            conclusion = f"Reject null hypothesis: sample mean is significantly greater than population mean at alpha={alpha}."
        else:
            conclusion = f"Reject null hypothesis: sample mean is significantly less than population mean at alpha={alpha}."
    else:
        conclusion = f"Fail to reject null hypothesis: no significant difference between sample mean and population mean at alpha={alpha}."

    return z_stat, p_value, conclusion


if __name__ == "__main__":
    sample_mean = 52
    population_mean = 50
    population_std = 8
    sample_size = 36
    alpha = 0.05
    alternative = 'two-sided'

    z_stat, p_value, conclusion = one_sample_z_test(sample_mean, population_mean, population_std, sample_size, alpha, alternative)

    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(conclusion)
```

```python
#A-4
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import sqrt

def two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    """
    Perform a two-tailed Z-test and visualize decision regions.

    Parameters:
    - sample_mean: Mean of the sample
    - population_mean: Known population mean
    - population_std: Known population standard deviation
    - sample_size: Number of observations in the sample
    - alpha: Significance level

    Returns:
    - z_stat: Computed Z statistic
    - p_value: Corresponding p-value
    """
    std_error = population_std / sqrt(sample_size)
    z_stat = (sample_mean - population_mean) / std_error

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    z_crit = norm.ppf(1 - alpha / 2)

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.3, label='Rejection region (left)')
    plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.3, label='Rejection region (right)')

    plt.fill_between(x, 0, y, where=((x > -z_crit) & (x < z_crit)), color='green', alpha=0.3, label='Acceptance region')

    plt.axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f'Observed Z = {z_stat:.3f}')

    plt.title('Two-tailed Z-test Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_stat, p_value

if __name__ == "__main__":
    sample_mean = 105
    population_mean = 100
    population_std = 15
    sample_size = 30
    alpha = 0.05

    z_stat, p_value = two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha)
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at alpha={alpha}. The sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at alpha={alpha}. No significant difference detected.")
```

```python
#A-5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import sqrt

def visualize_type1_type2_errors(pop_mean, pop_std, sample_size, alpha, true_mean_alt):
    """
    Calculate and visualize Type 1 and Type 2 errors for a one-sample Z-test.

    Parameters:
    - pop_mean: Population mean under null hypothesis
    - pop_std: Population standard deviation (known)
    - sample_size: Number of observations in sample
    - alpha: Significance level (Type 1 error rate)
    - true_mean_alt: True mean under alternative hypothesis

    Returns:
    - type1_error: The alpha value (significance level)
    - type2_error: Calculated beta value (Type 2 error)
    """
    std_error = pop_std / sqrt(sample_size)

    z_crit_low = norm.ppf(alpha / 2)
    z_crit_high = norm.ppf(1 - alpha / 2)

    crit_val_low = pop_mean + z_crit_low * std_error
    crit_val_high = pop_mean + z_crit_high * std_error

    beta_low = norm.cdf(crit_val_high, loc=true_mean_alt, scale=std_error) - norm.cdf(crit_val_low, loc=true_mean_alt, scale=std_error)
    type2_error = beta_low

    x_min = pop_mean - 4 * std_error
    x_max = pop_mean + 4 * std_error
    x = np.linspace(x_min, x_max, 1000)

    h0_pdf = norm.pdf(x, loc=pop_mean, scale=std_error)
    h1_pdf = norm.pdf(x, loc=true_mean_alt, scale=std_error)

    plt.figure(figsize=(12, 7))
    plt.plot(x, h0_pdf, label='H0: Null Distribution', color='blue')
    plt.plot(x, h1_pdf, label='H1: Alternative Distribution', color='orange')

    plt.fill_between(x, 0, h0_pdf, where=(x <= crit_val_low), color='red', alpha=0.3, label='Type 1 Error (α)')
    plt.fill_between(x, 0, h0_pdf, where=(x >= crit_val_high), color='red', alpha=0.3)

    plt.fill_between(x, 0, h1_pdf, where=((x > crit_val_low) & (x < crit_val_high)), color='purple', alpha=0.3, label='Type 2 Error (β)')

    plt.axvline(crit_val_low, color='black', linestyle='--', label=f'Critical Value Low: {crit_val_low:.2f}')
    plt.axvline(crit_val_high, color='black', linestyle='--', label=f'Critical Value High: {crit_val_high:.2f}')

    plt.title('Visualization of Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type 1 error (α): {alpha:.4f}")
    print(f"Type 2 error (β): {type2_error:.4f}")
    print(f"Power (1 - β): {1 - type2_error:.4f}")

    return alpha, type2_error


if __name__ == "__main__":
    pop_mean = 100
    pop_std = 15
    sample_size = 30
    alpha = 0.05
    true_mean_alt = 110

    visualize_type1_type2_errors(pop_mean, pop_std, sample_size, alpha, true_mean_alt)
```

```python
#A-6
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05, equal_var=True):
    """
    Perform an independent two-sample T-test and interpret the results.

    Parameters:
    - sample1: List or array of sample data from group 1
    - sample2: List or array of sample data from group 2
    - alpha: Significance level (default 0.05)
    - equal_var: Assume equal population variances (default True)

    Returns:
    - t_stat: Calculated T statistic
    - p_value: Corresponding p-value
    - interpretation: String interpretation of the test result
    """
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=equal_var)

    if p_value < alpha:
        interpretation = (f"Reject the null hypothesis at alpha={alpha}. "
                          "There is a significant difference between the two groups.")
    else:
        interpretation = (f"Fail to reject the null hypothesis at alpha={alpha}. "
                          "There is no significant difference between the two groups.")

    return t_stat, p_value, interpretation

if __name__ == "__main__":
    np.random.seed(42)
    group1 = np.random.normal(loc=20, scale=5, size=30)
    group2 = np.random.normal(loc=22, scale=5, size=30)

    alpha = 0.05
    t_stat, p_value, interpretation = independent_t_test(group1, group2, alpha)

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(interpretation)
```

```python
#A-7
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test(before, after, alpha=0.05):
    """
    Perform a paired T-test and visualize the results.

    Parameters:
    - before: List or array of measurements before treatment
    - after: List or array of measurements after treatment
    - alpha: Significance level (default 0.05)

    Returns:
    - t_stat: T statistic
    - p_value: P-value
    - interpretation: Hypothesis test result
    """

    t_stat, p_value = ttest_rel(before, after)

    if p_value < alpha:
        interpretation = (f"Reject the null hypothesis at alpha={alpha}. "
                          "There is a significant difference between the paired groups.")
    else:
        interpretation = (f"Fail to reject the null hypothesis at alpha={alpha}. "
                          "There is no significant difference between the paired groups.")

    plt.figure(figsize=(10, 6))
    x = np.arange(len(before))
    plt.plot(x, before, marker='o', label='Before', color='blue')
    plt.plot(x, after, marker='o', label='After', color='green')
    for i in range(len(before)):
        plt.plot([x[i], x[i]], [before[i], after[i]], 'k--', alpha=0.5)

    plt.title('Paired Sample T-Test: Before vs After')
    plt.xlabel('Subject Index')
    plt.ylabel('Measurement')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return t_stat, p_value, interpretation


if __name__ == "__main__":
    np.random.seed(0)
    before = np.random.normal(loc=100, scale=10, size=20)
    after = before + np.random.normal(loc=-5, scale=5, size=20)

    t_stat, p_value, interpretation = paired_t_test(before, after)

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(interpretation)
```

```python
#A-8
import numpy as np
from scipy.stats import norm, t
from math import sqrt

def simulate_data(mu=50, sigma=10, sample_size=30, delta=2):
    """
    Simulate a sample from a normal distribution.

    Parameters:
    - mu: population mean
    - sigma: population std deviation
    - sample_size: number of samples
    - delta: shift from population mean (effect size)

    Returns:
    - sample: numpy array of simulated sample data
    - population_mean: the population mean
    - population_std: the known population std deviation (for Z-test)
    """
    np.random.seed(42)
    sample = np.random.normal(loc=mu + delta, scale=sigma, size=sample_size)
    return sample, mu, sigma

def perform_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.
    """
    sample_mean = np.mean(sample)
    n = len(sample)
    std_error = pop_std / sqrt(n)
    z_stat = (sample_mean - pop_mean) / std_error
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    conclusion = "Reject" if p_value < alpha else "Fail to reject"
    return z_stat, p_value, conclusion

def perform_t_test(sample, pop_mean, alpha=0.05):
    """
    Perform a one-sample T-test.
    """
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)
    std_error = sample_std / sqrt(n)
    t_stat = (sample_mean - pop_mean) / std_error
    df = n - 1
    p_value = 2 * (1 - t.cdf(abs(t_stat), df))

    conclusion = "Reject" if p_value < alpha else "Fail to reject"
    return t_stat, p_value, conclusion

if __name__ == "__main__":
    sample, pop_mean, pop_std = simulate_data(mu=50, sigma=10, sample_size=30, delta=2)

    z_stat, z_p, z_conclusion = perform_z_test(sample, pop_mean, pop_std)

    t_stat, t_p, t_conclusion = perform_t_test(sample, pop_mean)

    print("\n--- Comparison of Z-test and T-test Results ---")
    print(f"Sample Mean: {np.mean(sample):.4f}")
    print(f"Known Population Mean (H0): {pop_mean}")
    print(f"Sample Size: {len(sample)}")

    print(f"\nZ-test:\n  Z-statistic = {z_stat:.4f}\n  P-value = {z_p:.4f}\n  Conclusion: {z_conclusion} H0")

    print(f"\nT-test:\n  T-statistic = {t_stat:.4f}\n  P-value = {t_p:.4f}\n  Conclusion: {t_conclusion} H0")
```

```python
#A-9
import numpy as np
from scipy.stats import t

def confidence_interval_mean(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean using the T-distribution.

    Parameters:
    - sample: array-like, list or numpy array of sample data
    - confidence: confidence level (e.g., 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): confidence interval tuple
    - margin_of_error: margin added/subtracted from the mean
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    se = sample_std / np.sqrt(n)

    df = n - 1
    t_crit = t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_crit * se
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return (ci_lower, ci_upper), margin_of_error

if __name__ == "__main__":
    np.random.seed(42)
    sample_data = np.random.normal(loc=100, scale=15, size=25)

    ci, moe = confidence_interval_mean(sample_data, confidence=0.95)

    print(f"Sample Mean: {np.mean(sample_data):.2f}")
    print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
    print(f"Margin of Error: {moe:.2f}")
```

```python
#A-10
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(sample, confidence_level=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
    - sample: array-like, list or numpy array of sample values
    - confidence_level: confidence level (e.g., 0.95 for 95%)

    Returns:
    - margin_of_error: the margin of error value
    """
    n = len(sample)
    sample_std = np.std(sample, ddof=1)
    standard_error = sample_std / np.sqrt(n)
    df = n - 1

    t_crit = t.ppf((1 + confidence_level) / 2, df)

    margin_of_error = t_crit * standard_error
    return margin_of_error

if __name__ == "__main__":
    np.random.seed(1)
    sample_data = np.random.normal(loc=100, scale=12, size=30)

    confidence_level = 0.95
    moe = calculate_margin_of_error(sample_data, confidence_level)

    print(f"Sample Size: {len(sample_data)}")
    print(f"Confidence Level: {confidence_level * 100:.1f}%")
    print(f"Margin of Error: ±{moe:.2f}")
```

```python
#A-11
def bayes_theorem(prior_H, prob_D_given_H, prob_D_given_not_H):
    """
    Apply Bayes' Theorem to update belief based on new evidence.

    Parameters:
    - prior_H: P(H) = Prior probability of the hypothesis (e.g., having the disease)
    - prob_D_given_H: P(D|H) = Likelihood (test is positive given hypothesis is true)
    - prob_D_given_not_H: P(D|~H) = Probability of test being positive when hypothesis is false

    Returns:
    - posterior: P(H|D) = Updated probability of hypothesis given the data
    """

    prior_not_H = 1 - prior_H
    prob_D = (prob_D_given_H * prior_H) + (prob_D_given_not_H * prior_not_H)

    posterior = (prob_D_given_H * prior_H) / prob_D

    return posterior

if __name__ == "__main__":
    prior = 0.01

    sensitivity = 0.99
    false_positive_rate = 1 - 0.95

    posterior = bayes_theorem(prior, sensitivity, false_positive_rate)

    print(f"Posterior Probability of Having Disease Given Positive Test: {posterior:.4f}")
```

```python
#A-12
import pandas as pd
from scipy.stats import chi2_contingency


data = {'Product A': [30, 20],'Product B': [20, 30]}
contingency_table = pd.DataFrame(data, index=['Male', 'Female'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:")
print(contingency_table)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=['Male', 'Female'], columns=['Product A', 'Product B']))
print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

alpha = 0.05
if p < alpha:
    print("\n🔍 Conclusion: Reject the null hypothesis — variables are NOT independent.")
else:
    print("\n✅ Conclusion: Fail to reject the null hypothesis — variables are independent.")
```

```python
#A-13
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test from an observed contingency table.

    Parameters:
    - observed: 2D array-like (e.g., nested list, NumPy array, or DataFrame)

    Returns:
    - expected: 2D NumPy array of expected frequencies
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total

    return expected

if __name__ == "__main__":
    observed_data = [[30, 20],[20, 30]]

    observed_df = pd.DataFrame(observed_data, index=['Male', 'Female'], columns=['Product A', 'Product B'])

    expected = calculate_expected_frequencies(observed_data)
    expected_df = pd.DataFrame(expected, index=['Male', 'Female'], columns=['Product A', 'Product B'])

    print("Observed Frequencies:")
    print(observed_df)
    print("\nExpected Frequencies (if variables are independent):")
    print(expected_df.round(2))
```

```python
#A-14
import numpy as np
from scipy.stats import chisquare

observed = np.array([15, 22, 18, 20, 19, 26])

total_rolls = observed.sum()
expected = np.full(shape=6, fill_value=total_rolls / 6)

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Observed:", observed)
print("Expected:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("🔍 Reject the null hypothesis: The die is likely not fair.")
else:
    print("✅ Fail to reject the null hypothesis: The die appears fair.")
```
